## 1. Defensible Choices

Every decision in this investigation had a reason. Here they are, stated plainly.

**Choice: Analysis window — July to September 2023**
Derived mathematically from roll number digits (R=8), giving critical points at months 7 and 9. This is not arbitrary — it is reproducible and tied to a specific function. The window sits between two behavioural turning points of f(x), which is exactly where the commissioner's concern lies.

**Choice: Yellow Taxi as primary dataset**
Yellow taxi data has the richest schema — fare, tip, payment type, passenger count, distance. HVFHV data lacks tip and payment columns in comparable form. Green taxi volume is too small to draw borough-level conclusions. Yellow taxi is the most defensible choice for behavioural analysis.

**Choice: 7 cleaning rules**
Each rule removes physically impossible or analytically useless records. Zero-distance trips never moved. Negative-duration trips violate causality. Trips over 6 hours are almost certainly parked meters or data entry errors. The 94–97% retention rate confirms these rules are conservative, not aggressive.

**Choice: 8% fare increase as the policy simulation (Act 5)**
The magnitude is tied to R=8. The parameter — fare — is the TLC's primary lever. The 2022 TLC fare increase was 7.6%, making 8% historically grounded, not invented.

**Choice: Mann-Whitney U for tip comparison (Act 3)**
Tip percentages are not normally distributed — they are right-skewed with a spike at zero. A parametric t-test would be inappropriate. Mann-Whitney U makes no normality assumption.

## 2. Assumptions Questioned

At least once in every act, the analysis turned on itself.

**Assumption questioned: Peak hours make passengers more generous**
This is the commission's stated belief. The data shows tip rates are higher at peak hours — but so is the card payment rate. Since TLC only records tips for card payments, the tip difference may be a measurement artifact, not genuine generosity. The assumption may be mechanically true but behaviourally false.

**Assumption questioned: Weekend fare/mile drop reflects passenger behaviour**
The alternative — that weekend trips include more airport runs with flat-rate metering — is equally consistent with the data. Without trip purpose data, the behavioural explanation cannot be confirmed. This was acknowledged, not hidden.

**Assumption questioned: The 7-day cycle is structural**
The periodogram shows a dominant 7-day frequency. But with only 92 days of data, the periodogram has low resolution. A 7-day spike could reflect the calendar structure of the data itself, not an independent property of the system. The conclusion is probable, not certain.

**Assumption questioned: The fare increase will reduce volume**
The elasticity estimate of -0.4 is borrowed from literature, not estimated from this data. NYC taxi demand may be less elastic than the literature suggests — captive users (tourists, airport travellers, late-night riders) have few alternatives. The simulation shows a range of outcomes precisely because this assumption is uncertain.

## 3. Limits of the Conclusions

What this analysis can say — and what it cannot.

**Can say:**
- The data quality is high: 94–97% of records survive conservative cleaning
- The city has a consistent weekly rhythm: the 7-day cycle is real and stable
- Trip volume shows no significant trend across Jul–Sep 2023: the system is stable at the monthly scale
- Credit card dominates payment (78.5%): cash is declining as a share
- An 8% fare increase would raise revenue under all elasticity assumptions
- Tip compression is a plausible unintended consequence of a fare increase

**Cannot say:**
- Whether drivers are earning more or less than previous years (no multi-year comparison)
- Whether passengers are satisfied or dissatisfied (no survey data)
- Whether the patterns observed in July–September hold in other seasons
- Whether cash tippers behave differently from card tippers (cash tips are invisible)
- Whether the fare increase would cause passengers to shift to HVFHV (no cross-platform elasticity)
- Whether the 'zones of stress' identified are caused by competition, geography, or driver choice

## 4. Final Verdict — Real Signal or Convincing Noise?

The commissioner asked one question: **is the system functioning normally, or quietly drifting into inefficiency?**

The evidence, taken together:

| Signal | Real or Noise? | Confidence |
|--------|---------------|------------|
| Weekly 7-day cycle | **Real** — confirmed by periodogram | High |
| Stable monthly volume | **Real** — linear trend not significant | High |
| Peak-hour tip premium | **Uncertain** — card rate confound unresolved | Medium |
| Weekend fare/mile drop | **Real** — but cause uncertain (behaviour vs route) | Medium |
| Distance-tip correlation | **Noise** — r near zero, flat-rate norm likely | High |
| Fare increase tip compression | **Plausible** — not confirmed, modelled only | Low-Medium |

**Verdict:**

The NYC yellow taxi system in July–September 2023 is **not in measurable decline**. Volume is stable. Data quality is high. The weekly rhythm is structural and consistent.

However, the system shows **early signs of quiet drift**:
- Passengers tip by habit, not by journey — driver effort is not rewarded proportionally
- Card payment dominance means cash tippers are invisible to analysis — a growing blind spot
- The fare/mile efficiency gap between peak and off-peak hours suggests the system is not uniformly productive

**The signs are real. But they are not yet loud enough to call a crisis.**

What the commissioner should watch:
- Tip rate trends over multiple years (not just one quarter)
- HVFHV market share growth by borough
- Driver utilisation rates (trips per active hour) — not available in this dataset

*The system is holding. But it is not holding equally everywhere, for everyone, at every hour. That is the honest conclusion.*
